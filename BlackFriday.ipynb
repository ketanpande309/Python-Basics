# <b>Pandas GroupBy operation</b>

```python
#use group by on top of gender column
df1.groupby("Gender")
```

```python
#See the groups
df1.groupby("Gender").groups
```

```python
#get groups based on gender and age combination
df1.groupby(['Gender','Age'])
```

```python
df1.groupby(['Gender','Age']).groups
```

```python
#apply aggregation function sum with GroupBy
df1.groupby("Gender").sum()
```

```python
#use np.sum
df1.groupby("Gender").aggregate(np.sum)
```

```python
#get total purchased amount
df1.groupby("Gender").Purchase.aggregate(np.sum)
```

```python
#get sum as well as mean
df1.groupby("Gender").Purchase.aggregate([np.sum,np.mean])
```

```python
#We can also apply function on top of group
df1[df1.groupby("Gender").Purchase.apply(lambda x:x>10000)]
```
