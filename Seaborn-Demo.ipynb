# <b>Scatter Plot</b>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

<b>We can generate scatter plot using two functions</b>

<b>1.relplot</b>

<b>2.scatterplot</b>

```python
#Draw scatter plot on total_bill and tip column
plt.figure(figsize=(30,50))
sns.relplot(x='total_bill',y='tip',data=data)
```

```python
#use hue parameter to show the smoker column
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker')
```

```python
#use style parameter to show the sex column
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker',style='sex')
```

```python
#use size parameter to show the size column which will increase the size of points accordingly
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker',style='sex',size='size')
```

```python
#plot using scatter plot function
plt.figure(figsize=(7,7))
sns.scatterplot(x='total_bill',y='tip',data=data,hue='smoker',style='sex',size='size')
```

# <b>Strip plot and Swarm plot</b>

<b>Both these plots can be plot using catplot function or using individual name as well</b>

<b>For these plots, we will use 1 Categorical and 1 numerical variable</b>

```python
#If we use catplot, we have to give kind parameter to specify whether to draw strip plot or swarm plot, by default it will take strip plot
sns.catplot(x='day',y='tip',data=data)
```

```python
sns.catplot(x='day',y='tip',data=data, hue='sex')
```

```python
sns.catplot(x='day',y='tip',data=data,kind='swarm')
```

```python
sns.catplot(x='day',y='tip',data=data,kind='swarm',hue='sex')
```

```python
sns.stripplot(x='day',y='tip',data=data,hue='sex')
```

```python
sns.swarmplot(x='day',y='tip',data=data,hue='sex')
```

<b>Difference between strip and swarm plot is that in swarm plot we can get the data more distributed form</b>

# <b>Box plot and Violin plot</b>

<b>Both these plots can be plot using catplot function or using individual name as well</b>

<b>We will use these plots on numerical variable to generate five number summary(min,25th percentile, median(50th percentile), 75th percentile and max)<b>

<b>Also we will use it based on category</b>

```python
sns.catplot(x='total_bill',data=data,kind='box')
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='box',hue='sex')
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x='day',y='total_bill',data=data,hue='sex')
```

```python
plt.figure(figsize=(8,8))
sns.violinplot(x='total_bill',data=data)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin', hue='sex',data=data)
```

```python
sns.catplot(x='day',y='total_bill',kind='violin', hue='sex',split=True,data=data)
```

<b>The advantage of violin plot over box plot is that, in violin plot along with statistical data we get how normally distributed the data is</b>

# <b>Bar plot and Count plot</b>

<b>Both these plots can be plot using catplot function or using individual name as well<b>

<b>In bar plot, we can check central tendency (mean,median, mode, std, etc) of a data</b>

<b>In count plot, we can check the frequency of categorical data</b>

```python
sns.catplot(x='day',y='total_bill', kind='bar',data=data)
```

```python
sns.catplot(x='day',y='total_bill',estimator=np.median, kind='bar',data=data)
```

```python
sns.catplot(x='day',y='total_bill',estimator=np.median, kind='bar',hue='sex',data=data)
```

```python
sns.barplot(x='day',y='total_bill',estimator=np.median,hue='sex',data=data)
```

```python
sns.catplot(x='sex',kind='count',data=data)
```

```python
sns.catplot(x='sex',kind='count',hue='smoker',data=data)
```

```python
sns.countplot(x='sex',hue='smoker',data=data)
```

# <b>Heat map</b>

<b>In heatmap we can analyse the correlation between two variables in dataset, i.e how related they are to each other</b>

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x = pd.pivot_table(data=flights,values='passengers',columns='month',index='year',aggfunc='sum')
```

```python
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(x,cbar=False, cmap='winter',annot=True,fmt='d')
```

# <b>Cluster maps</b>

<b>Cluster maps are heat maps only forming the clusters, i.e same type of row occurs together and same type of column together</b>

```python
plt.figure(figsize=(10,10))
sns.clustermap(x)
```

```python
sns.clustermap(x, annot=True, metric='correlation', fmt='d')
```

```python
sns.clustermap(x, annot=True, metric='correlation', fmt='d', row_cluster=False)
```

```python
sns.clustermap(x, annot=True, metric='correlation', fmt='d', col_cluster=False)
```

# <b>Dist plot</b>

<b>Distplot is used to do univariate analysis on continuous/numerical data</b>

<b>We can plot distplot using 3 plots</b>

<b>1.Histogram</b>

<b>2.KDE</b>

<b>3.Rug plot</b>

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'],kde=False)
```

```python
sns.distplot(titanic['age'], hist=False)
```

```python
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

<b>We can distplot to analyze whether which age group people died max or which age group people survived max</b>

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```

<b>So here, we can analyze that most of the people who died are in the age group of 25-40 and probability of surviving is lesser than not surviving. Also most of the children survived
